```python
#task 1
import numpy as np
import matplotlib.pyplot as plt
a = 0.1
b = -0.5
c = 25.0

time = np.linspace(0, 10, 100)


temperature = a * time**2 + b * time + c


max_temperature = max(temperature)
max_time_index = np.argmax(temperature)
max_time = time[max_time_index]


plt.figure(figsize=(8, 6))
plt.plot(time, temperature, label=f"{a}t^2 + {b}t + {c}")
plt.scatter([max_time], [max_temperature], color='red', label='Maximum Temperature')
plt.title('Temperature Prediction Using Quadratic Model')
plt.xlabel('Time (days)')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()

print(f"Maximum temperature of {max_temperature} occurs at {max_time} days.")
```

```python
#task 2
import cmath
import numpy as np
import matplotlib.pyplot as plt


a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))


discriminant = cmath.sqrt(b**2 - 4*a*c)


x1 = (-b - discriminant) / (2*a)
x2 = (-b + discriminant) / (2*a)

print("Roots:")
print("x1 =", x1)
print("x2 =", x2)


x = np.linspace(-10, 10, 400)

y = a*x**2 + b*x + c

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"{a}x^2 + {b}x + {c}")
plt.scatter([x1.real, x2.real], [0, 0], color='red', label='Roots')
plt.title('Quadratic Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example historical data (time in days, temperature in Celsius)
time = np.array([0, 1, 2, 3, 4, 5])  # days
temperature = np.array([20, 22, 25, 24, 23, 21])  # Celsius

# Fit a quadratic curve to the data
coefficients = np.polyfit(time, temperature, 2)  # Fit a quadratic polynomial

# Extract coefficients
a = coefficients[0]
b = coefficients[1]
c = coefficients[2]

# Print the quadratic equation
print(f"The quadratic equation is: y = {a}t^2 + {b}t + {c}")

# Predict future temperature at day 6
future_time = 6
predicted_temperature = a * future_time**2 + b * future_time + c
print(f"Predicted temperature at day {future_time}: {predicted_temperature} Celsius")

# Plot the data and the fitted curve
plt.figure()
plt.scatter(time, temperature, label='Data points')
plt.plot(time, a*time**2 + b*time + c, color='red', label='Quadratic fit')
plt.xlabel('Time (days)')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Prediction Using Quadratic Model')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#task 3
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, list):
    a = list[0]
    b = list[1]
    c = list[2]

    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)
    listval = [[40, -3, 10]]

    for idx, line in enumerate(listval):
        temp = quadratic_model(time_values, line)
        plt.plot(time_values, temp, label=f'Set {idx + 1} {line}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Hard-Coded in List)')
    plt.show()

if __name__ == "__main__":
    main()
```
